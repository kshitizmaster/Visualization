```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

# Donut plot

```python
plt.figure(figsize=(9,9))
tickets = [48 , 30 , 20 , 15]
labels = ['Low' , 'Medium' , 'High' , 'Critical']
colors = ['#8BC34A','#D4E157','#FFB300','#FF7043']
plt.pie(tickets, labels = labels, colors = colors, startangle = 45)
my_circle = plt.Circle((0,0),0.7, color = 'white')
p= plt.gcf()
p.gca().add_artist(my_circle)
plt.legend(loc = 'lower right')
plt.show()
```

```python
# Changing background color
fig = plt.figure(figsize = (9,9))
fig.patch.set_facecolor('#DADADA')
tickets = [48 , 30 , 20 , 15]
labels = ['Low' , 'Medium' , 'High' , 'Critical']
colors = ['#8BC34A','#D4E157','#FFB300','#FF7043']
plt.pie(tickets, labels = labels, colors = colors)
my_circle = plt.Circle((0,0), 0.7, color = '#DADADA')
p = plt.gcf()
p.gca().add_artist(my_circle)
plt.legend()
plt.show()
```

#### Explode Slice in Donut Chart

```python
fig = plt.figure(figsize = (9,9))
fig.patch.set_facecolor('#DADADA')
tickets = [48 , 30 , 20 , 15]
labels = ['Low' , 'Medium' , 'High' , 'Critical']
colors = ['#8BC34A','#D4E157','#FFB300','#FF7043']
plt.pie(tickets, labels= labels, colors = colors, explode = [0.03,0.03,0.03,0.03], autopct = '%1.1f%%', pctdistance = 0.85, startangle = 45)
my_Circle=  plt.Circle((0,0), 0.7, color = '#DADADA')
p = plt.gcf()
p.gca().add_artist(my_Circle)
plt.legend()
plt.show()
```

```python
fig = plt.figure(figsize=(9,9))
fig.patch.set_facecolor('#DADADA')
tickets = [48 , 30 , 20 , 15]
total = np.sum(tickets)

def val_per(x): 
    return '{:.2f}%\n({:.0f})'.format(x, total*x/100)

labels = ['Low' , 'Medium' , 'High' , 'Critical']
colors = ['#8BC34A','#D4E157','#FFB300','#FF7043']
plt.pie (tickets , labels= labels , colors= colors , startangle=45 , autopct=val_per, pctdistance=0.85 ,explode=[0,0 , 0 , 0])
my_circle=plt.Circle( (0,0), 0.7, color='#DADADA') # Adding circle at the centre
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```
