```python

import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#graph Styling
plt.style.use('seaborn-darkgrid')
```

```python
#Simple Bar Chart
id1 = np.arange(1,10)
score = np.arange(20,110,10)
plt.bar(id1, score)
plt.xlabel('student IDs')
plt.ylabel('Score')
plt.show()
```

```python
# Changing color of the bar chart
id1 = np.arange(1,10)
score = np.arange(20,110,10)
plt.figure(figsize = (8,8)) #setting the figure size
ax = plt.axes()
ax.set_facecolor('#ECF0F1') #setting the background color
plt.bar(id1, score)
plt.xlabel(r'$Student$ IDs')
plt.ylabel(r'$Score$')
plt.show()
```

```python
#Plotting multiple sets of data
x1= [1,3,5,7] 
x2=[2,4,6,8]
y1 = [7,7,7,7]
y2= [17,18,29,40]
plt.figure(figsize= (10,10))
ax = plt.axes()
ax.set_facecolor('white')
plt.bar(x1, y1, color = '#42B300', label='First')# plotting the first set of Data
plt.bar(x2, y2, color = '#94E413', label = 'Second') #plotting the second set of Data
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title('$Bar$ $Chart$')
plt.legend()
plt.show()
```

```python
# Horizontal Bar Chart
Age = [28,33,43,45,57]
Name = ["Asif", "Steve", 'John', "Ravi", "Basit"]
plt.barh(Name, Age, color = 'yellowgreen')
plt.xlabel('$Name$')
plt.ylabel('$Age$')
plt.show()
```

```python
# Changing the width of Bars
num1 = np.array([1,3,5,7,9])
num2 = np.array([2,4,6,8,10])
plt.bar(num1, num1**2, color = '#FF6F00', label= "First", width = 0.2)
plt.bar(num2, num2**2, color = '#FFB300' , label = "Second", width = 0.2)
plt.legend()
plt.show()
```

```python
# Displaying values at the top of vertical bars
num1 = np.array([1,3,5,7,9])
num2 = np.array([2,4,6,8,10])
plt.figure(figsize = (10,6))
plt.bar(num1, num1**2, width=0.3 , color = '#FF6F00', label = 'First')
plt.bar(num2, num2**2, width=0.3 , color = '#FFB300', label = "Second")
for x, y in zip(num1, num1**2):
    plt.text(x, y+0.5, '%d' %y, ha = 'center', va = 'bottom')
for x, y in zip(num2, num2**2):
    plt.text(x, y +0.5, '%d' %y, ha = 'center', va = 'bottom')
plt.legend()
plt.show()
```

```python
x = np.arange(1,21)
plt.figure(figsize=(16,8))
y1 = np.random.uniform(0.1,0.7,20)
y2 = np.random.uniform(0.1,0.7,20)
ax = plt.axes()
ax.set_facecolor('white')
plt.bar(x, y1, label = 'First', edgecolor = 'white')
plt.bar(x, -y2, label = 'Second', edgecolor='white')
for x, y in zip(x, y1):
    plt.text(x, y+0.05, '%.2f'% y, ha = 'center', va = 'bottom', fontsize = 10)

plt.legend()
plt.xlabel('$X$')
plt.ylim(-1,1)
plt.xlim(0,23)
plt.show()
```

## Stacked Vertical Bar

```python
plt.style.use('seaborn-darkgrid')
x1= ['Asif','Basit','Ravi','Minil'] 
y1= [17,18,29,40]
y2 = [20,21,22,23]
plt.figure(figsize = (10,6))
plt.bar(x1, y1, color = '#FF6F00', label = 'Open Tickets', width = 0.5)
plt.bar(x1, y2, color = '#FFB300', label = 'Closed Tickets', bottom = y1, width = 0.5)
plt.legend()
plt.xlabel('$Names$')
plt.title ('$Bar $ $ Chart$')
plt.show()
```

```python
plt.style.use('seaborn-darkgrid')
x1= ['Asif','Basit','Ravi','Minil'] 
y1= np.array([17,18,29,40])
y2 =np.array([20,21,22,23])
y3 =np.array([5,9,11,12])
plt.figure(figsize=(5,7))
plt.bar(x1, y1, color = '#FF6F00', label = 'Open Tickets', width = 0.5)
plt.bar(x1, y2, bottom = y1, color = '#FFB300', label = 'Closed Tickets', width = 0.5)
plt.bar(x1, y3, bottom = y1+y2, color = '#F7DC6F', label = 'Cancelled Tickets', width = 0.5)
plt.legend()
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title ('$Bar $ $ Chart$')
plt.show()
```

## Grouped Bar Chart

```python
plt.figure(figsize = (7,9))
#set the bar width
barWidth = 0.25
# set height of bar
y1= np.array([17,18,29,40])
y2 =np.array([20,21,22,23])
y3 =np.array([5,9,11,12])
#set the position of bar on x axis
pos = np.arange(len(y1))
pos1 = [x + barWidth for x in pos]
pos2 = [x+ barWidth for x in pos1]

# Make the plot
plt.bar(pos, y1, width = barWidth, label = 'Open', color = '#FBC02D')
plt.bar(pos1,y2,  width = barWidth , label ='Closed', color = '#F57F17')
plt.bar(pos2, y3,  width = barWidth, label = 'Cancelled', color = '#E65100')

# Add xticks on the middle of the group bars
plt.xlabel('Assignee', fontweight='bold')
plt.ylabel('Number of Tickets', fontweight='bold')
plt.xticks([i + barWidth for i in range(len(y1))],['Asif', 'Basit', 'Ravi', 'Minil'])
plt.legend()
plt.show()
```

## Stacked Horizontal Bar

```python
plt.style.use('seaborn-darkgrid')
x1= ['Asif','Basit','Ravi','Minil'] 
y1= [17,18,29,40]
y2 = [20,21,22,23]
plt.figure(figsize=(8,5))
plt.barh(x1, y1, label = "Open Tickets",color = '#FF6F00')
plt.barh(x1, y2, label = "Closed Tickets", left = y1 , color = '#FFB300')
plt.legend()
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title ('$Bar $ $ Chart$')
plt.show()
```

## Displaying values in Bar Charts

```python
# Displaying values in the stacked vertical bars using plt.text()
plt.style.use('seaborn-darkgrid')
x1= ['Asif','Basit','Ravi','Minil'] 
y1= [17,18,29,40]
y2 = [20,21,22,23]
plt.figure(figsize=(5,7))
plt.bar(x1, y1, label = 'Open Tickets', color = '#FF6F00', width = 0.5)
plt.bar(x1, y2, label  = 'Closed Tickets', color = '#FFB300', bottom = y1, width = 0.5)

for x, y in zip(x1, y1):
    plt.text(x, y-10,'%d' %y, ha = 'center', va = 'bottom')
for x, y, z in zip(x1, y2, y1):
    plt.text(x, y+z-10, '%d' %y, ha = 'center', va = 'bottom')

plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title ('$Bar $ $ Chart$')
    
plt.legend()
```

```python
# Displaying values in the stacked horizontal bars using plt.text()
plt.style.use('seaborn-darkgrid')
x1= ['Asif','Basit','Ravi','Minil'] 
y1= [17,18,29,40]
y2 = [20,21,22,23]
plt.figure(figsize=(8,5))
plt.barh(x1, y1, color = '#FF6F00', label = 'Open Ticket')
plt.barh(x1, y2, color = '#FFB300', label = 'Closed Ticket', left = y1)

for x, y in zip(x1, y1):
    plt.text(y-10, x, '%d' %y, fontsize = 20, ha = 'center', va = 'center')
for x, y, z in zip(x1, y2, y1):
    plt.text(y+z-10, x, '%d'%y, fontsize = 20, ha = 'center', va = 'center')
    
plt.legend()
```

```python
# Displaying values at the top of the Grouped Bar Chart using plt.text()
plt.figure(figsize=(7,9))
#setting the bandwidth
bandWidth = 0.25

# set height of bar
y1= np.array([17,18,29,40])
y2 =np.array([20,21,22,23])
y3 =np.array([5,9,11,12])

#setting the x-coordinate
pos = np.arange(len(y1))
pos1 = [x+ bandWidth for x in pos]
pos2 = [x + bandWidth for x in pos1]

plt.bar(pos, y1, width = bandWidth, label = 'Open Ticket', color = '#FBC02D')
plt.bar(pos1, y2, width = bandWidth , label = 'Closed Ticket', color = '#F57F17')
plt.bar(pos2, y3 , width = bandWidth, label = 'Cancalled Ticket', color = '#E65100')

# Add xticks on the middle of the group bars
plt.xlabel('Assignee', fontweight='bold')
plt.ylabel('Number of Tickets', fontweight='bold')
plt.xticks([x+ bandWidth for x in range(len(y1))], ['Asif', 'Basit', 'Ravi', 'Minil'])

for x , y in zip(pos, y1):
    plt.text(x, y, '%d'%y, ha = 'center', va ='bottom')

for x , y in zip(pos1, y2):
    plt.text(x, y, '%d'%y, ha = 'center', va ='bottom')

for x , y in zip(pos2, y3):
    plt.text(x, y, '%d'%y, ha = 'center', va ='bottom')


plt.title ('$Grouped $  $ Bar $  $ Chart$')

# Create legend & Show graphic
plt.legend()
plt.show()
```
